```python
import utils 
import numpy as np 
import pandas as pd 
import warnings 
import csv 
import os

warnings.filterwarnings('ignore')
```

```python
csv_file_paths = os.listdir("../chatgpt0613-csv")
csv_file_paths = ["../chatgpt0613-csv/"+ file 
                   for file  in csv_file_paths]
print(csv_file_paths)

acc = pd.DataFrame(columns=["Q1", "Q2", "Q3", "Q4"])

for csv_file_path in csv_file_paths:
    
    l = [0, 0, 0, 0]
    
    number_type = ("prime" if "prime" in csv_file_path else "composite")
    df = pd.read_csv(csv_file_path)
    
    for i in range(len(df)):
      for j in range(4):
        idx = f"Q{j+1}"
        result1 = utils.return_PrimeOrComposite(df.loc[i][idx], j) 
        if result1 == number_type:
          l[j] += 1
    
    for i in range(4):      
      l[i] /= len(df) 
      l[i] *= 100
    acc.loc[csv_file_path[len("../chatgpt0613-csv/"):]] = l
          
acc
    
```
